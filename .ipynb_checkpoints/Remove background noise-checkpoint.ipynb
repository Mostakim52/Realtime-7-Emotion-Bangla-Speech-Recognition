```python
import librosa
import librosa.display
import noisereduce as nr
import numpy as np
import scipy.io.wavfile as wav

# Load audio file
def remove_noise(input_audio, output_audio):
    rate, data = wav.read(input_audio)

    # If stereo, convert to mono
    if len(data.shape) > 1:
        data = np.mean(data, axis=1)

    # Apply noise reduction
    reduced_noise = nr.reduce_noise(y=data, sr=rate)

    # Save the denoised audio
    wav.write(output_audio, rate, reduced_noise.astype(np.int16))

# Example usage
remove_noise(r"E:\Downloads\KUET Bangla Emotional Speech (KBES) Dataset\Angry (High)\4-2-1-001.wav", "output_clean.wav")
```
